Here's an example of how to use, with a matplotlib example of random data generated by Numpy

Importing needed libraries

```python
import matplotlib.pyplot as plt
import numpy as np
from parsecpt import create_colormap, parse_cpt_string #Functions to parse the strings and create probable color maps for matplotlib
from colorgen import ColorToolApp
```

Creating A Panoply CPT Generator instance.

```python
CPTGen = ColorToolApp() #Creating an Instance of the tool, this variable later will be set as the CPT Content
```


Here's a picture of the output, I modified the colors to make my own color map, you can also click on the color buttons and modify them manually rather than dragging, after I click Save CPT As class __str__ the variable we set the class instance as which is CPTGen here will be identified as the CPT Content, so if I do print(CPTGen) I'd get the CPT Content.

```python
print(CPTGen)
```

Now We Use the 2 Functions we imported from parsecpt to turn it into a format that makes it usable in matplotlib for our test

```python
cpt_colors = parse_cpt_string(CPTGen.__str__())
colormap = create_colormap(cpt_colors)
```

Now we create a plot with random data from numpy, the output will differ each time you run this.

```python
data = np.random.rand(10, 10) * 100
plt.imshow(data, cmap=colormap)
plt.colorbar()
plt.title("Map with CPT Colormap")
plt.show()
```
